```python
# 데이터 프레임의 컬럼의 개수가 많을 경우 중간 정보가 생략된다.
# 컬럼 이름을 확인하고 싶을 때 컬럼목록을 가져와서 출력을 해야 한다.
for c1 in df1.columns:
    print(c1)
```

```python
# 각 컬럼별로 결측치 개수를 확인한다.
# isna() : 각각의 자리의 값이 결측치라면 true, 아니라면 false로 되어 있는 데이터 프레임을 얻을 수 있다.
# sum() : 각 컬럼별 총합을 계산할 수 있다. 이때 True는 1, False는 0으로 취급하여 총합을 구한다. 
df1.isna(). sum()
```

```python
# 각 컬럼별 결측치를 제외한 데이터의 개수 파악
# notna : 결측치가 아닌 값은 True, 결측치는 False로 되어 있는 데이터 프레임을 생성한다.
# sum :  
df1.notna().sum()
```

```python
# 각 컬렴별 결측치 개수 데이터를 가져온다.
a1 = df1.isna().sum()
# 결측치가 있는 컬럼만 가져온다
a2 = a1[a1>0]

a1 = df1.dtypes
for a3 in a2.index:
    print(f'{a3} : {a2[a3]}')
```

### 데이터의 종류
- 범위형 : 연속성을 가지고 있는 값을 의미한다. 최소나 최대가 있을 수 있다(나이, 키, 몸무게 등등)
- 범주형(레이블형, 카테고리형) : 값의 종류가 정해져 있고 연속성을 가지고 있지 않는 데이터(혈액형, 성별, 지하철 노선 번호 등등)||

```python
# 요약통계정보
# count : 결측치를 제외한 데이터의 개수
# mean : 결측치를 제외한 평균
# std : 결측치를 제외한 표준편차
# min : 결측치를 제외한 최소값
# 25% : 결측치를 제외한 25% 위치의 값
# 50% : 결측치를 제외한 50% 위치의 값
# 75% : 결측치를 제외한 최대값
df1.describe()
```

```python
# 위에서 보면 데이터프레임이 가지고 있는 모든 숫자컬럼의 데이터를 모아 보여준다.
# 이에 범주형 데이터가 숫자 컬럼이면 이 결과에 포함된다.
# 따라서 범위형 데이터만 모아서 살펴봐야 하낟.
df1[['Age', 'SibSp', 'Parch']]
```

```python
# astype : 데이터 프레임이나 Series가 가진 데이터들을 원하는 타입으로 변경한 데이터르 프레임을 가져올 수 있다.
# 이렇게 하여 범주형 데이터들을 모두 문자열로 변환한 다음 요약 통계정보를 살펴보면 된다.
# count : 결측치를 제외한 데이터의 개수
# unique : 값의 종류 개수
# top : 가장 많이 나온 데이터
# freq : 가장 많이 나온 데이터가 몇 번 나왔는지
df1[['Sex', 'Pclass']].astype('str').describe()
```

```python
# 범위형 데이터의 요약 통계 정보에서 min과 max를 확인하여 
# 데이터 명세서에 있는 정상범위 안에 모든 값들이 존재하는지 확인해야 한다.
# 혹은 나중에 배울 통계량을 구하는 함수를 이용해 최소와 최대값을 확인해도 된다.
df1[['Age', 'SibSp', 'Parch']].describe().loc[['min', 'max']]
```

```python
# 원하는 컬럼을 추출하여 각 컬럼별로 어떠한 값이 몇 개씩 있는지 확인한다.
# 이것을 통해서 데이터 명세서에 나와있지 않는 값이 저장되어 있는지 확인한다.
df1['Survived'].value_counts().index
```
