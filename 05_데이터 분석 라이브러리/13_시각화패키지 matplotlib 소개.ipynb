- matplotlib는 파이썬에서 자료를 시각화하는 대표적인 시각화 라이브러리
- 다양한 그래프를 그릴 수 있도록 함수들이 제공되고 있다.
- 다른 시각화 라이브러리들은 matplotlib를 통해 구현된 시각화 라이브러리이다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 그래프 그리기를 위한 기본 설정
# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프 크기
plt.rcParams['figure.figsize'] = 12, 6

# 글자 크기
plt.rcParams['font.size'] = 14

# 적용한 폰트에 - 기호가 없다면 깨진다. 
# - 기호는 설정한 폰트의 영향을 받지 않도록 설정한다.
plt.rcParams['axes.unicode_minus'] = False
```

### 그래프 그리기
- 라인 플롯을 통해서 그래프를 그리는 기본 방법에 대해 살펴본다.
- matplotlib는 그래프의 종류에 따라 함수가 제공되며 원하는 그래프를 그릴 수 있는 함수를 사용하면 된다.

```python
# x축의 값과 y축의 값을 지정하여 그래프를 그려준다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
# 그래프를 그리는 함수들 중 일부는 축 하나의 값만 지정할 수 있는 것들도 있다.
# 이 때는 y축의 값을 지정하는 것이 된다.
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 일부 그래프를 그리는 함수들은 반드시 x와 y축의 값을 지정해야 하는 것들도 있다.
plt.scatter([1, 4, 9, 16])
plt.show()
```

### 그래프의 제목

```python
# 그래프의 제목을 설정한다.
plt.title('그래프의 제목')
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 제목의 위치를 설정한다.
# loc : left, right, center
plt.title('그래프의 제목', loc = 'left')
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 제목의 세로 위치를 설정한다.
# y : 제목의 세로 위치를 설정한다. 
# 0이면 그래프의 아래쪽이 된다.
# 1은 그래프의 y축 높이만큼 된다.
# 기본은 None으로 되어 있으며, None은 1 위치에 해당된다. 

plt.title('그래프의 제목', loc = 'center', y = -0.15)
plt.plot([1, 4, 9, 16])
plt.show()
```

### 그래프의 스타일 지정

```python
plt.title('그래프의 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], '')
plt.show()
```

- 그래프의 스타일은 문자열로 지정한다.
- 문자열 양식은 '색상 마커종류 선종류' 형태로 작성한다(r = color, s = type of marker , -- = type of line)
---
- 색상
- base color 사용 (b,g,r,c,m,y,w,k)
- https://matplotlib.org/stable/gallery/color/named_colors.html
---
- 마커
- https://matplotlib.org/stable/api/markers_api.html
---
- 선 종류
- '-' : 실선
- '--' : dash
- '-.' : 점과 dash
- ':' : 점선

```python
plt.title('마젠타별모양점선')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'm*:')
plt.show()
```

### 상세설정
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 색상
- mew : 마커 선 굵기
- mfc : 마커 내부 색상

```python
plt.title('그래프의 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='r', lw=5, ls=':', 
        marker='o', ms=15, mec='g', mew=5, mfc='b')
plt.show()
```

### 그리드 설정
- matplotlib 옛날 버전에서는 그리드가 기본적으로 나왔으나 현재는 그렇지 않음. 

```python
plt.title('그래프의 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])


# 그리드가 나오게 한다.
plt.grid()
# 만약 grid가 기본으로 나오는 옛날버전일 경우 그리드가 안나오게 하는 코드
# plt.grid(False)
plt.show()
```

```python
# 하나의 좌표 평면에 여러 개의 그래프를 그린다.
# plt.show()를 호출하기 전까지 그리는 그래프들은 모두
# 하나의 좌표 평면에 나온다.
# 주피터에서는 plt.show() 호출하지 않는 경우
# 하나의 셀 안에서 그린 모든 그래프는 하나의 좌표 평면에 나타난다.
plt.title('그래프의 제목')

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.plot([10, 20, 30, 40], [16, 9, 4, 1])

plt.show()
```

```python
# plt.show() 를 호출한 다음 그리는 그래프들은 다른 좌표평면에 그려진다.
plt.title('그래프의 제목')

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()

plt.plot([10, 20, 30, 40], [16, 9, 4, 1])
plt.show()
```

```python
# subplot : 좌표 평면을 가로, 세로로 나눠 각각의 영역에 좌표 평면을 하나씩 배치하여 그래프를 그릴수 있다.

# 2행 2열의 1번째
# 행의 개수, 열의 개수, 몇 번째 인가.
plt.subplot(2, 2, 1)
plt.title('1행 1열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 2행 2열의 2번째
plt.subplot(2, 2, 2)
plt.title('1행 2열')
plt.plot([10, 20, 30, 40], [16, 9, 4, 1])

# 2행 2열의 3번째
plt.subplot(2, 2, 3)
plt.title('2행 1열')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# 2행 2열의 4번째
plt.subplot(2, 2, 4)
plt.title('2행 2열')
plt.plot([10, 20, 30, 40], [16, 9, 4, 1])

# 전체 타이틀
# y : 제목의 세로 높이(0은 제일 밑에, 1을 제일 위)
plt.suptitle('Sub Plot', y = 1.1)

# 그래프의 크기나 위치 등을 보기 좋게 자동으로 설정해준다.
# plt.tight_layout()

# 직접설정
plt.subplots_adjust(hspace=0.4)

plt.show()
```

### 범례

```python
plt.title('그래프의 제목')

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], label = '선1')
plt.plot([10, 20, 30, 40], [16, 9, 4, 1], label = '선2')

# 범례를 표시한다.
# loc : 범례의 위치, 생략시 0
# 0 : best, 가장 최적의 위치를 자동으로 선정함.
# 1 : 우상단(*), 2: 좌상단, 3 : 좌하단, 4 : 우하단, 5 : 중앙우측, 6 : 중앙좌측, 7 : 중앙우측, 8 : 중앙하단, 9: 중앙상단, 10 : 정중앙
plt.legend(loc=9)

plt.show()
```

```python
plt.title('그래프의 제목')

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], label = '선1')
plt.plot([10, 20, 30, 40], [16, 9, 4, 1], label = '선2')

# bbox_to_anchor : 범례를 좌표평면 밖으로 빼고 싶을 때 사용한다.
plt.legend(bbox_to_anchor = (1.01, 1))

plt.show()
```
