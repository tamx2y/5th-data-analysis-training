# 버블 정렬
- 인접한 값들끼리 비교하여 위치를 교환하는 방식으로 정렬을 수행한다.
- 교환이 한번도 이루어지지 않았다면 정렬이 완료된 것으로 판단한다.

![](doc_images/버블정렬1.jpg)


```python
def bubble_sort(data_list) :
    """
        data_list
            정렬할 데이터를 가지고 있는 리스트
    """
    # 반복한다.
    while True :
        # 교환이 이루어졌는지 확인을 위한 변수에 False값을 넣어준다.
        check_change = False
        # 데이터의 수 - 1 만큼 반복한다.
        for idx in range(len(data_list) - 1) :
            # 현재 위치의 변수의 값과 다음 위치의 변수의 값을 비교한다.
            # 만약 다음 위치의 값이 작으면
            if data_list[idx] > data_list[idx + 1] :
                # 두 변수의 값을 교환한다
                data_list[idx], data_list[idx + 1] = data_list[idx + 1], data_list[idx]
                # 교환이 이루어 졌는지 확인을 위한 변수에 True를 넣어준다.
                check_change = True
            print(data_list)
        print("=" * 30)
        # 만약 교환이 이루어지지 않았다면 반복을 중단한다.
        if check_change == False :
            break
```

```python
a1 = [10, 2, 7, 4, 1, 5, 8, 6, 3, 9]
bubble_sort(a1)
a1
```
