# Linked List
- 노드라는 단위로 데이터를 관리하는 개념
- 노드 : 1개 이상의 데이터를 하나로 묶은 개념
- Linked List는 관리할 값(객체)와 다음 노드의 주소값 혹은 이름으로 구성된다.
- 현재 노드에서 다음 노드가 무엇인지 파악할 때 노드가 관리하는 다음 노드의 주소값 혹은 이름으로 파악한다.
- head node
  - 기본적으로 존재하는 노드이고, 첫 번째 값(객체)를 관리하는 노드의 주소값 혹은 이름을 가지고 있는 노드이다.
  - 데이터를 추가, 삭제, 수정, 가져오기 등의 작업을 할 때 head node부터 순차적으로 접근한다.

![single linked list](doc_images/single_linked_list.jpg)

```python
# Single Linked List 클래스
class SingleLinkedList :

    def __init__(self) :
        # 데이터를 담을 딕셔너리
        self.linked_list = {
            'head' : [None, None]
        }
        # head node의 이름
        self.head_name = 'head'
        # 새로운 노드의 이름에 붙힐 숫자값
        self.node_count = 0

    # 데이터 추가하기
    def addData(self, value) :
        # 제일 마지막 노드의 이름을 파악한다.
        # 현재 노드의 노드의 이름
        current_node_name = self.head_name
        while  True :
            # 현재 노드에서 다음 노드의 이름을 가져온다.
            next_node_name = self.linked_list[current_node_name][1]
            # 만약 다음 노드가 없다면 while문을 종료한다.
            if next_node_name == None :
                break
            else :
                # 현재 노드의 이름을 담는 변수에 다음 노드의 이름을 담아준다.
                current_node_name = next_node_name
                
        # 새로운 노드를 만들어준다.
        new_node_name = f'node_{self.node_count}'
        self.node_count = self.node_count + 1
        new_node = [value, None]
        
        # 제일 마지막 노드가 관리하는 다음 노드의 이름을 새로운 노드의 이름을 넣어준다
        # 딕셔너리에 새로운 노드를 담아준다.
        self.linked_list[new_node_name] = new_node
        # 제일 마지막 노드가 관리하는 다음 노드의 이름을 새로운 노드의 이름으로
        # 넣어준다.
        self.linked_list[current_node_name][1] = new_node_name

        # 데이터 모두 출력하기
    def showData(self) :
        # 현재 노드의 이름을 헤드 노드의 이름으로 설정한다.
        current_node_name = self.head_name
        # 마지막 노드를 만날 때 까지 반복한다.
        while True:
            # 현재 노드가 head 노드가 아니라면 관리하는 값을 출력한다.
            if current_node_name != self.head_name : 
                print(self.linked_list[current_node_name][0])
            # 현재 노드가 관리하는 다음 노드의 이름이 None이면 중단한다.
            if self.linked_list[current_node_name][1] == None : 
                break
            # 현재 노드가 관리하는 다음 노드의 이름이 None이 아니면 현재 노드의 이름을 담는 변수에 다음 노드의 이름을 담아준다.
            else : 
                current_node_name = self.linked_list[current_node_name][1]
                
    # 순서를 통해 데이터 가져오기
    def getDataByPosition(self, position) :
        # 현재 노드의 이름을 헤더 노드의 이름으로 설정해준다.
        current_node_name = self.head_name
        # 순서값을 관리할 변수에 0을 넣어준다.
        current_position = 0
        
        # 만약 헤더 노드가 관리하는 다음 노드의 이름이 없다면 '저장된 데이터가 없습니다'를 출력한다.
        if self.linked_list[self.head_name][1] == None : 
            print('저장된 데이터가 없습니다')
            return

        # 반복한다.
        while True:
           # 다음 노드의 이름은 현재 노드의 이름으로 담아준다.
            current_node_name = self.linked_list[current_node_name][1]
            # 현재 노드의 위치값하고 찾을 노드의 위치값이 같으면 값을 출력한다.
            if current_position == position :
                print(self.linked_list[current_node_name][0])
                break
            # 현재 노드의 위치값하고 찾을 노드의 위치값이 다를 경우 
            else:
                # 만약 다음 노드의 이름이 None이면 순서값을 잘못 입력하였다고 출력하고 중단한다.
                if self.linked_list[current_node_name][1] == None :
                    print('순서값을 잘못 입력하였습니다')
                    break
                # 만약 다음 노드의 이름이 None이 아니면 현재 노드의 이름에 다음 노드의 이름을 넣어주고 순서값을 1 증가한다.
                else: 
                    current_position = current_position + 1
                    
       # 값을 통해서 데이터의 순서값을 가져온다.
    def getPositionByValue(self, value) :
        # 현재 노드의 이름을 헤더 노드의 이름으로 설정해준다.
        current_node_name = self.head_name

        # 현재 노드의 순서값을 0으로 설정한다.
        current_position = 0

        # 만약 헤더 노드가 관리하는 다음 노드의 이름이 없다면
        # 저장된 데이터가 없습니다를 출력한다.
        if self.linked_list[self.head_name][1] == None :
            print('저장된 데이터가 없습니다')
            return

        # 반복한다
        while True :
            # 다음 노드의 이름을 현재 노드의 이름으로 담아준다.
            current_node_name = self.linked_list[current_node_name][1]
            
            # 현재 노드가 관리하는 값과 찾고자 하는 값이 같으면 현재 노드의 순서값을 출력하고 중단한다.
            if self.linked_list[current_node_name][0] == value :
                print(f'{value}의 위치는 {current_position} 입니다')
                break

            # 만약 현재 노드의 값과 찾고자 하는 값이 다르다면
            else :
                # 만약 다음 노드의 이름이 None이면 '값을 찾지 못했습니다'를 출력하고 중단한다.
                if self.linked_list[current_node_name][1] == None :
                    print(f'{value}값을 찾지 못했습니다')
                    break
                # 만약 다음 노드의 이름이 None이 아니면 순서값을 1 증가시킨다.
                else :
                    current_position = current_position + 1
                
    # 데이터 삭제하기
    def deleteDataByPosition(self, position) :
        # 현재 노드의 이름을 헤더 노드의 이름으로 설정해준다.
        current_node_name = self.head_name
        # 순서값을 관리할 변수에 0을 넣어준다. 
        current_position = 0
        # 만약 헤더 노드가 관리하는 다음 노드의 이름이 없다면 '저장된 데이터가 없습니다'를 출력하고 중단한다.
        if self.linked_list[self.head_name][1] == None :
            print('저장된 데이터가 없습니다')
            return
            
        # 반복한다. 
        while True : 
            # 이전 노드의 이름에 현재 노드의 이름을 담아준다. 
            prev_node_name = current_node_name
            # 현재 노드의 이름에 다음 노드의 이름을 담아준다.
            current_node_name = self.linked_list[current_node_name][1]
            
            # 현재 노드의 위치값하고 찾을 노드의 위치값이 같다면,
            if current_position == position : 
                # 현재 노드가 관리하는 다음 노드의 이름을 가져온다.
                next_node_name = self.linked_list[current_node_name][1]
                # 이전 노드가 관리하는 다음 노드의 이름을 현재 노드가 관리하는 다음 노드의 이름으로 변경해준다.
                self.linked_list[prev_node_name][1] = next_node_name
                # 현재 노드를 삭제한다.
                del self.linked_list[current_node_name]
                # 중단한다. 
                break
                
            # 현재 노드의 위치값과 찾을 노드의 위치값이 다를 경우
            else : 
                # 만일 다음 노드의 이름이 None이면 순서값을 잘못 입력하였다고 출력하고 중단한다.
                if self.linked_list[current_node_name][1] == None : 
                    print('순서값을 잘못 입력하였습니다')
                    break
                # 만약 다음 노드의 이름이 None이 아니면 순서값 1을 증가시킨다.
                else :
                    current_position = current_position + 1

    # 데이터 삽입하기
    def insertDataByPosition(self, value, position) :
        # 현재 노드의 이름을 헤더 노드의 이름으로 설정해준다.
        current_node_name = self.head_name
        # 순서값을 관리할 변수에 0을 넣어준다.
        current_position = 0
        # 만약 헤더 노드가 관리하는 다음 노드의 이름이 없다면 '저장된 데이터가 없습니다'를 출력하고 중단한다.
        if self.linked_list[self.head_name][1] == None :
            print('저장된 데이터가 없습니다')
            return
            
        # 반복한다.
        while True : 
            # 이전 노드의 이름에 현재 노드의 이름을 담아준다.
            prev_node_name = current_node_name
            # 현재 노드의 이름에 다음 노드의 이름을 담아준다.
            current_node_name = self.linked_list[current_node_name][1]
            # 현재 노드의 위치값이 찾을 노드의 위치값과 같다면
            if current_position == position : 
                # 새로운 노드를 생성한다.
                new_node_name = f'node_{self.node_count}'
                self.node_count = self.node_count + 1
                new_node = [value, None]
                self.linked_list[new_node_name] = new_node
                # 새로운 노드가 관리할 다음 노드의 이름을 현재 노드의 이름으로 설정한다.
                new_node[1] = current_node_name
                # 이전 노드가 관리할 다음 노드의 이름을 새로운 노드의 이름으로 설정한다. 
                self.linked_list[prev_node_name][1] = new_node_name
                # 중단한다. 
                break

         # 현재 노드의 위치값하고 찾을 노드의 위치값이 다를 경우
            else :
                # 만약 다음 노드의 이름이 None이면 순서값을 잘못 입력하였다고 출력하고 중단한다.
                if self.linked_list[current_node_name][1] == None :
                    print('순서값을 잘못 입력하였습니다')
                    break
                # 만약 다음 노드의 이름이 None이 아니면 순서값 1을 증가시킨다.
                else :
                    current_position = current_position + 1

 # 데이터 수정하기 
    def setDataByPosition(self, value, position) :
       # 현재 노드의 이름을 헤더 노드의 이름으로 설정해준다.
        current_node_name = self.head_name
        # 순서값을 관리할 변수에 0을 넣어준다.
        current_position = 0
        # 만약 헤더 노드가 관리하는 다음 노드의 이름의 없다면
        # 저장된 데이터가 없습니다를 출력하고 중단한다.
        if self.linked_list[self.head_name][1] == None :
            print('저장된 데이터가 없습니다')
            return

        # 반복한다.
        while True :
            # 현재 노드의 이름에 다음 노드의 이름을 담아준다.
            current_node_name = self.linked_list[current_node_name][1]
            # 현재 노드의 위치값하고 찾을 노드의 위치 값이 같다면..
            if current_position == position :
                # 현재 노드에 새로운 값을 넣어준다.
                self.linked_list[current_node_name][0] = value
                # 중단한다.
                break

            # 현재 노드의 위치값하고 찾을 노드의 위치값이 다를 경우
            else :
                # 만약 다음 노드의 이름이 None이면 순서값을 잘못 입력하였다고 출력하고 중단한다.
                if self.linked_list[current_node_name][1] == None :
                    print('순서값을 잘못 입력하였습니다')
                    break
                # 만약 다음 노드의 이름이 None이 아니면 순서값 1을 증가시킨다.
                else :
                    current_position = current_position + 1
```

```python
single_linked_list = SingleLinkedList()

single_linked_list.getPositionByValue(100)

single_linked_list.addData(100)
single_linked_list.addData(200)
single_linked_list.addData(300)

# single_linked_list.showData()
single_linked_list.setDataByPosition(2000, 1)

single_linked_list.showData()
```
